```python
#Set-up
from selenium import webdriver
import pandas as pd 
```

```python
#Inicialización, abrir el website
driver = webdriver.Chrome('C:\webdrivers\chromedriver.exe')
driver.get('https://pokemondb.net/pokedex/all')
driver.maximize_window()
```

```python
#Aceptar la política de privacidad del website
button = driver.find_element_by_class_name('gdpr-accept')
try:
    button.click()
except:
    pass
```

```python
#Definir las variables de los datos que queremos obtener del website
#nombre
pokemon_nombre = driver.find_elements_by_xpath('//tbody/tr/td[2]')
#tipo
pokemon_tipo = driver.find_elements_by_xpath('//tbody/tr/td[3]')
#total
pokemon_total = driver.find_elements_by_xpath('//tbody/tr/td[4]')
#vida
pokemon_vida = driver.find_elements_by_xpath('//tbody/tr/td[5]')
#ataque
pokemon_ataque = driver.find_elements_by_xpath('//tbody/tr/td[6]')
#defensa
pokemon_defensa = driver.find_elements_by_xpath('//tbody/tr/td[7]')
#ataque especial
pokemon_ataque_especial = driver.find_elements_by_xpath('//tbody/tr/td[8]')
#defensa especial
pokemon_defensa_especial = driver.find_elements_by_xpath('//tbody/tr/td[9]')
#velocidad
pokemon_velocidad = driver.find_elements_by_xpath('//tbody/tr/td[10]')
```

```python
#Lista vacía a la que vamos a agregar todos los pokemones con sus datos
pokemon_resultados = []
```

```python
#Diccionario que combina todos los pares valor/clave generados por el paso n°4
for i in range(len(pokemon_nombre)):
    pokemon_diccionario = {'Nombre':pokemon_nombre[i].text,
                       'Tipo': pokemon_tipo[i].text,
                       'Total':pokemon_total[i].text,
                       'Vida': pokemon_vida[i].text,
                       'Ataque':pokemon_ataque[i].text,
                       'Defensa': pokemon_defensa[i].text,
                       'Ataque Especial': pokemon_ataque_especial[i].text,
                       'Defensa Especial': pokemon_defensa_especial[i].text,
                       'Velocidad':pokemon_velocidad[i].text}
    
    pokemon_resultados.append(pokemon_diccionario)
```

```python
#Crear resumen de nuestros resultados utilizando panda
dataframe_pokemon = pd.DataFrame(pokemon_resultados)
```

```python
#Mostrar el resumen de nuestros resultados
dataframe_pokemon
```

```python
#Quitar el "/n" que aparece entre los valores "tipo" y "nombre" si el pokemon posee más de uno
dataframe_pokemon['Tipo'] = dataframe_pokemon['Tipo'].apply(lambda x: x.replace ('\n', ', '))
dataframe_pokemon['Nombre'] = dataframe_pokemon['Nombre'].apply(lambda x: x.replace ('\n', '/ '))
```

```python
#Mostrar el resumen de nuestros resultados con las correcciones del paso anterior
dataframe_pokemon
```

```python
#Exportar nuestro dataframe hacia un archivo Excel
dataframe_pokemon.to_excel('Pokedex_final.xlsx', index=False)
```
